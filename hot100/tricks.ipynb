# [Leetcode 136. 只出现一次的数字](https://leetcode.cn/problems/single-number/submissions/717939700/?envType=study-plan-v2&envId=top-100-liked)

```cpp
// Solution in C++
class Solution {
public:
    int singleNumber(vector<int>& nums) {
        int ans = nums[0];
        for (int i = 1; i < nums.size(); i++) {
            ans = ans ^ nums[i];
        }
        return ans;
    }
};
```

````python
# Solution in Python
class Solution:
    def singleNumber(self, nums):
        ans = nums[0]
        for i in range(1, len(nums)):
            ans = ans ^ nums[i]
        return ans
````

# [Leetcode 169. 多数元素](https://leetcode.cn/problems/majority-element/?envType=study-plan-v2&envId=top-100-liked)

```cpp
class Solution {
public:
    int majorityElement(vector<int>& nums) {
        unordered_map<int,int> counts;
        int n = nums.size();
        for (int num : nums) {
            counts[num]++;
            if (counts[num] > n / 2) {
                return num;
            }
        }
        return -1;
    }
};
```

````python
from typing import List

class Solution:
    def majorityElement(self, nums: List[int]) -> int:
        counts = {}
        n = len(nums)
        for num in nums:
            counts[num] = counts.get(num, 0) + 1
            if counts[num] > n // 2:
                return num
        return -1
````

# [Leetcode 75. 颜色分类](https://leetcode.cn/problems/sort-colors/description/?envType=study-plan-v2&envId=top-100-liked)

```cpp
class Solution {
public:
    void sortColors(vector<int>& nums) {
        int count[3] = {0, 0, 0};
        for (int num : nums) {
            count[num]++;
        }
        int index = 0;
        for (int color = 0; color < 3; color++) {
            while (count[color] != 0) {
                nums[index] = color;
                index++;
                count[color]--;
            }
        }
    }
};
```

````python
from typing import List

class Solution:
    def sortColors(self, nums: List[int]) -> None:
        """
        Do not return anything, modify nums in-place instead.
        """
        count = [0, 0, 0]
        for num in nums:
            count[num] += 1
        index = 0
        for i in range(3):
            for _ in range(count[i]):
                nums[index] = i
                index += 1
````

# [Leetcode 31. 下一个排列](https://leetcode.cn/problems/next-permutation/?envType=study-plan-v2&envId=top-100-liked)

```cpp
class Solution {
public:
    void nextPermutation(vector<int>& nums) {
        int n = nums.size();
        int i = n - 2;
        while (i >= 0 && nums[i] >= nums[i + 1]) i--;
        if (i >= 0) {
            int j = n - 1;
            while (j > i && nums[j] <= nums[i]) j--;
            swap(nums[i], nums[j]);
        }
        reverse(nums.begin() + i + 1, nums.end());
    }
};
```

````python
from typing import List

class Solution:
    def nextPermutation(self, nums: List[int]) -> None:
        """
        Do not return anything, modify nums in-place instead.
        """
        n = len(nums)
        # 1. 从右向左找第一个升序对 (i, i+1)
        i = n - 2
        while i >= 0 and nums[i] >= nums[i + 1]:
            i -= 1
        
        # 2. 如果找到了这样的 i，则从右向左找第一个大于 nums[i] 的数
        if i >= 0:
            j = n - 1
            while j > i and nums[j] <= nums[i]:
                j -= 1
            # 交换 nums[i] 和 nums[j]
            nums[i], nums[j] = nums[j], nums[i]
        
        # 3. 反转 i+1 到末尾（原为降序，反转后变为升序）
        left, right = i + 1, n - 1
        while left < right:
            nums[left], nums[right] = nums[right], nums[left]
            left += 1
            right -= 1
````

# [Leetcode 287. 寻找重复数](https://leetcode.cn/problems/find-the-duplicate-number/?envType=study-plan-v2&envId=top-100-liked)

```cpp
class Solution {
public:
    int findDuplicate(vector<int>& nums) {
        // 阶段1：快慢指针找相遇点
        int slow = nums[0];
        int fast = nums[0];
        do {
            slow = nums[slow];
            fast = nums[nums[fast]];
        } while (slow != fast);

        // 阶段2：找环入口（重复数字）
        slow = nums[0];
        while (slow != fast) {
            slow = nums[slow];
            fast = nums[fast];
        }
        return slow;
    }
};
```

````python
from typing import List

class Solution:
    def findDuplicate(self, nums: List[int]) -> int:
        # 阶段1：快慢指针找相遇点
        slow = fast = 0
        while True:
            slow = nums[slow]          # 慢指针走一步
            fast = nums[nums[fast]]    # 快指针走两步
            if slow == fast:
                break

        # 阶段2：找环入口
        slow = 0
        while slow != fast:
            slow = nums[slow]
            fast = nums[fast]
        return slow
````
