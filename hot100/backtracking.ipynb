```cpp
class Solution {

    void backtrack(vector<vector<int>> &res, vector<int> &nums, int start) {
        if (start == nums.size()) {
            res.push_back(nums);
            return;

    }
```

# [Leetcode17-电话号码的数字组合](https://leetcode.cn/problems/letter-combinations-of-a-phone-number/description/?envType=study-plan-v2&envId=top-100-liked)

```cpp
class Solution {

    std::unordered_map<char, std::string> phoneMap = {

        for (const auto &ch : letters) {
            combinatin.push_back(ch);
            backtrack(digits, index + 1, combination, res);
            combination.pop_back();
        }
    }
public:
    vector<string> letterCombinations (string digits) {
        vector<string> res;
```

````python
# 使用 Pyhton 求解
from typing import List

class Solution:
    def letterCombinations(self, digits: str) -> List[str]:
        res: List[str] = []
        if digits is None:
            return res

        phone_map = {
            '2': "abc",
            '3': "def",
            '4': "ghi",
            '5': "jkl",
            '6': "mno",
            '7': "pqrs",
            '8': "tuv",
            '9': "wxyz",
        }

        def backtrack(index: int, combination: List[str]):
            if len(combination) == len(digits):
                res.append(''.join(combination))
                return
            
            current_char: str = digits[index]
            letters: str = phone_map[current_char]

            for letter in letters:
                combination.append(letter)
                backtrack(index + 1, combination)
                combination.pop()
            
        backtrack(0, [])
        return res
````

# [Leetcode39-组合总和](https://leetcode.cn/problems/combination-sum/?envType=study-plan-v2&envId=top-100-liked)

```cpp
// Soluiton in C++
class Solution {
private:
    void backtrack (
        vector<int> &path,
        vector<vector<int>> &res,
        vector<int> &candidates,
        int target,
        int start
    ) {
        if (target == 0) {
            res.push_back(path);
            return;
        }

        if (target < 0) {
            return;
        }

        for (int i = start; i < candidates.size(); i++) {
            path.push_back(candidates[i]);
            backtrack(path, res, candidates, target - candidates[i], i);
            path.pop_back();
        }
    }
public:
    vector<vector<int>> combinationSum (vector<int> &candidates, int target) {
        vector<vector<int>> res;
        vector<int> path;
        backtrack(path, res, candidates, target, 0);
        return res;
    }
};
```

````python
# Solution in Python
from typing import List

class Solution:
    def combinationSum(self, candidates: List[int], target: int) -> List[list[int]]:
        res: List[List[int]] = []
        path: List[int] = []
        def backtrack(start: int, remaining: int):
            if remaining == 0:
                res.append(path[:])
                return 
            if remaining < 0:
                return
            
            for i in range(start, len(candidates)):
                path.append(candidates[i])
                backtrack(i, remaining - candidates[i]) # 注意这里是 remaining - candidates[i] !
                path.pop()

        backtrack(0, target)
        return res
````

# [Leetcode22-括号生成](https://leetcode.cn/problems/generate-parentheses/description/?envType=study-plan-v2&envId=top-100-liked)

```cpp
// Solution in C++

using std::string;

class Solution {

/**
 * 注意有效括号的条件
 * 1. 左右括号的数量是相等的
 * 2. 在任何位置，左括号的数量大于等于右括号的数量
 */

private:
    void backtrack(
        vector<string> &res,
        string current, // current 是当前的字符串是什么
        int open, // open 代表的是左括号的数量
        int close, // close 是右括号的数量
        int n // n 代表的是括号的对数
    ){
        if (current.size() == 2 * n) {
            res.push_back(current);
            return;
        }
        
        // 选择 1，添加左括号
        if (open < n) {
            backtrack(res, current + '(', open + 1, close, n);
        }

        // 选择 2，添加右括号
        if (close < open) {
            backtrack(res, current + ')', open, close + 1, n);
        }
        
    }

public:
    vector<string> generateParenthesis(int n) {
        vector<string> res;
        backtrack(res, "", 0, 0, n);
        return res;
    }
};
```

````python
# Solution in Python

from typing import List

class Solution:
    def generateParenthesis(self, n: int) -> List[str]:
        res: List[str] = []
        def backtrack(current: str, open: int, close: int):
            if len(current) == 2 * n:
                res.append(current)
                return
            
            if open < n:
                backtrack(current + '(', open + 1, close)
            
            if open > close:
                backtrack(current + ')', open, close + 1)

        backtrack("", 0, 0)
        return res
````

# [Leetcode79-单词搜索](https://leetcode.cn/problems/word-search/description/?envType=study-plan-v2&envId=top-100-liked)

```cpp
// Solution in C++
class Solution {
public:
    bool exist(vector<vector<char>> &board, string word) {
        int rows = board.size();
        int cols = board[0].size();
        vector<vector<bool>> visited(rows, vector<bool>(cols, false));
        vector<pair<int, int>> directions = {
            {1, 0},
            {-1, 0},
            {0, 1},
            {0, -1},
        };

        function<bool(int, int, int)> dfs = [&](int i, int j , int k) -> bool {
            // 成功匹配的话，返回所有的字符
            if (k == word.size()) {
                return true;
            }

            // 失败的条件：越界、已经访问、不匹配
            if (i < 0 || i >= rows || j < 0 || j >= cols || visited[i][j] || board[i][j] != word[k]) {
                return false;
            }
            visited[i][j] = true;
            for (auto [dx, dy] : directions) {
                if (dfs(i + dx, j + dy, k + 1)) {
                    return true;
                }
            }
            visited[i][j] = false;
            return false;
        };

        for (int i = 0; i < rows; i++) {
            for (int j = 0; j < cols; j++) {
                if (board[i][j] == word[0] && dfs(i, j, 0)) {
                    return true;
                }
            }
        }
        return false;
    }
};
```


````python
# Solution in Python

from typing import List

class Solution:
    def exist(self, board: List[List[str]], word: str) -> bool:
        rows: int = len(board)
        cols: int = len(board[0])

        visited: List[List[bool]] = [[False] * cols for _ in range(rows)]

        def dfs(i: int, j: int, k: int) -> bool:
            """
            Args:
                i 和 j 指的是当前位置
                k 指的是要匹配的字符的索引
            """

            if k == len(word):
                return True
            
            # 失败条件
            if i < 0 or i >= rows or j < 0 or j >= cols or visited[i][j] or board[i][j] != word[k]:
                return False
            
            visited[i][j] = True

            if dfs(i + 1, j, k + 1) or dfs(i - 1, j, k + 1) or dfs(i, j - 1, k + 1) or dfs(i, j + 1, k + 1):
                return True
            
            visited[i][j] = False
            return False
        
        for i in range(rows):
            for j in range(cols):
                if board[i][j] == word[0] and dfs(i, j, 0):
                    return True
                
        return False
````

# [Leetcode131-分割回文串](https://leetcode.cn/problems/palindrome-partitioning/?envType=study-plan-v2&envId=top-100-liked)

```cpp
// Solution in C++
class Solution {
private:
    bool isPalindrome (const string& word, int left, int right) {
        while (left < right) {
            if (word[left++] != word[right--]) {
                return false;
            }
        }
        return true;
    }

    void backtrack (
        const string &s,
        int start,
        vector<string> &path,
        vector<vector<string>> &res
    ) {
        if (start == s.size()) {
            res.push_back(path);
            return;
        }

        for (int end = start; end < s.size(); end++) {
            if (isPalindrome(s, start, end)) {
                path.push_back(s.substr(start, end - start + 1));
                backtrack(s, end + 1, path, res);
                path.pop_back();
            }

        }
    }
public:
    vector<vector<string>> partition(string s) {
        vector<vector<string>> res;
        vector<string> path;
        backtrack(s, 0, path, res);
        return res;
    }
};
```

````python
# Solution in Python

from typing import List

class Solution:
    def partition(self, s: str) -> List[List[str]]:
        res: List[List[str]] = []
        path: List[str] = []

        def is_palindrome(sub: str) -> bool:
            """
            判断一个字符串是否是回文的
            """
            return sub == sub[::-1]
        
        def backtrack(start: int) -> None:
            if (start == len(s)):
                res.append(path[:])
                return
            for end in range(start, len(s)):
                substring = s[start: end + 1]
                if is_palindrome(substring):
                    path.append(substring)
                    backtrack(end + 1)
                    path.pop()

        backtrack(0)
        return res
````

# [Leetcode51-N皇后](https://leetcode.cn/problems/n-queens/?envType=study-plan-v2&envId=top-100-liked)

```cpp
// Solution in C++
class Solution {
private:
    /**
     * 用来判断放在哪里是否合适，是否冲突
     */
    bool isValid (vector<vector<string> &board, int row, int col) {
        int n = board.size();

        // 检查同一个列
        for (int i = 0; i < row; i++) {
            if (board[i][col] == 'Q') {
                return false;
            }
        }

        // 检查左上的对角线
        for (int i = row - 1, int j = col - 1; i >= 0 && j >= 0; i--, j--) {
            if (board[i][j] == 'Q') {
                return false;
            }
        }

        // 检查右下对角线
        for (int i = row - 1, j = col + 1; i >= 0 && j < n; i--, j++) {
            if (board[i][j] == 'Q') {
                return false;
            }
        }

        return true;
    }

    void backtrack(vector<string> &board, int row, vector<vector<string>> &result) {
        int n = board.size();
        if (row == n) {
            result.push_back(board);
            return;
        }


        for (int col = 0; col < n; col++) {
            if (isValid(board, row, col)) {
                board[row][col] = 'Q';

                backtrack(board, row + 1, result);

                board[row][col] = '.';
            }
        }
    }
public:
    /**
     * 问题核心是如何进行冲突的检测
     */
     vector<vector<string>> solveNQueens (int n) {
        vector<vector<string>> res;
        vector<string> board(n, string(n, '.'));
        backtrack(board, 0, res);
        return res;
     }
};
```

````python
from typing import List

class Solution:
    def solveNQueens(self, n: int) -> List[List[str]]:
        res: List[List[str]] = []
        board = [['.' for _ in range(n)] for _ in range(n)]

        def is_valid(board: List[List[str]], row: int, col: int) -> bool:
            for i in range(row):
                if board[i][col] == 'Q':
                    return False
            
            i, j = row - 1, col - 1

            while i >= 0 and j >= 0:
                if board[i][j] == 'Q':
                    return False
                i -= 1
                j -= 1
            i, j = row - 1, col + 1
            while i >= 0 and j < n:
                if board[i][j] == 'Q':
                    return False
                i -= 1
                j += 1
            return True
        
        def backtrack(row: int) -> None:
            if row == n:
                solution = ["".join(row) for row in board]
                res.append(solution)
                return 
            
            for col in range(n):
                if is_valid(board, row, col):
                    board[row][col] = 'Q'

                    backtrack(row + 1)

                    board[row][col] = '.'

        backtrack(0)
        return res
````
