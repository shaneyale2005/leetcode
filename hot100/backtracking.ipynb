```cpp
// Cpp求解，第一种解法
class Solution {
private:
    void backtrack(vector<vector<int>> &res, vector<int> &nums, int start) {
        if (start == nums.size()) {
            res.push_back(nums);
            return;
        }

        for (int i = start; i < nums.size(); i++) {
            std::swap(nums[start], nums[i]);
            backtrack(res, nums, start + 1);
            std::swap(nums[start], nums[i]);
        }
        
    }
public:
    vector<vector<int>> permute(vector<int> &nums) {
        vector<vector<int>> res;
        backtrack(res, nums, 0);
        return res;
    }
};
```

```cpp
class Solution {
private:
    void backtrack(vector<vector<int>> &res, vector<int> &nums, vector<int> &path, vector<bool> &used) {
        if (path.size() == nums.size()) {
            res.push_back(path);
            return;
        }

        for (int i = 0; i < nums.size(); i++) {
            if (!used[i]) {
                used[i] = true;
                path.push_back(nums[i]);
                backtrack(res, nums, path, used);
                path.pop_back();
                used[i] = false;
            }
        }
    }
public:
    vector<vector<int>> permute(vector<int> &nums) {
        int n = nums.size();
        vector<vector<int>> res;
        vector<int> path;
        vector<bool> used(n, false);
        backtrack(res, nums, path, used);
        return res;
    }
};
```

````python
# 第一种解法

from typing import List

class Solution:
    def permute(self, nums: List[int]) -> List[List[int]]:
        res = []
        def backtrack(start: int):
            if start == len(nums):
                res.append(nums[:])
                return
            
            for i in range(start, len(nums)):
                nums[i], nums[start] = nums[start], nums[i]
                backtrack(start + 1)
                nums[i], nums[start] = nums[start], nums[i]
            
        backtrack(0)
        return res
````

````python
# 第二种解法
from typing import List

class Solution:
    def permute(self, nums: List[int]) -> List[List[int]]:
        res = []
        def backtrack(path, used):
            if len(path) == len(nums):
                res.append(path[:])
                return
            for i in range(len(nums)):
                if not used[i]:
                    used[i] = True
                    path.append(nums[i])
                    backtrack(path, used)
                    path.pop()
                    used[i] = False

        backtrack([], [False] * len(nums))
        return res
````

# [Leetcode78-子集](https://leetcode.cn/problems/subsets/description/?envType=study-plan-v2&envId=top-100-liked)

```cpp
// 继续使用回溯求解
class Solution {
private:
    void backtrack(
        vector<int> &nums,
        int start, vector<int> &path,
        vector<vector<int>> &res
    ) {
        res.push_back(path);

        for (int i = start; i < nums.size(); i++) {
            path.push_back(nums[i]);
            backtrack(nums, i + 1, path, res);
            path.pop_back();
        }
    }
public:
    vector<vector<int>> subsets(vector<int> &nums) {
        vector<int> path;
        vector<vector<int>> res;
        backtrack(nums, 0, path, res);
        return res;
    }
};
```

````python
# 使用回溯求解
from typing import List

class Solution:
    def subsets(self, nums: List[int]) -> List[List[int]]:
        result: List[List[int]] = []
        def backtrack(start, path):
            result.append(path[:])

            for i in range(start, len(nums)):
                path.append(nums[i])
                backtrack(i + 1, path)
                path.pop()
        backtrack(0, [])
        return result

'''
初始: []
├── 选择1: [1]
│   ├── 选择2: [1, 2]
│   │   └── 选择3: [1, 2, 3]
│   └── 选择3: [1, 3]
├── 选择2: [2]
│   └── 选择3: [2, 3]
└── 选择3: [3]
'''
````

# [Leetcode17-电话号码的数字组合](https://leetcode.cn/problems/letter-combinations-of-a-phone-number/description/?envType=study-plan-v2&envId=top-100-liked)

```cpp
// 使用 Cpp 求解
#include <vector>
#include <string>
#include <unordered_map>

class Solution {
private:
    std::unordered_map<char, std::string> phoneMap = {
        {'2', "abc"},
        {'3', "def"},
        {'4', "ghi"},
        {'5', "jkl"},
        {'6', "mno"},
        {'7', "pqrs"},
        {'8', "tuv"},
        {'9', "wxyz"},
    };

    void backtrack (
        const std::string &digits,
        int index,
        std::string &combination,
        std::vector<std::string> &res,
    ) {
        if (index == digits.size()) {
            res.push_back(combination);
            return;
        }

        // 获取当前数字对应的字母
        char digit = digits[index];
        std::string letters = phoneMap[digit];
        for (const auto &ch : letters) {
            combinatin.push_back(ch);
            backtrack(digits, index + 1, combination, res);
            combination.pop_back();
        }
    }
public:
    vector<string> letterCombinations (string digits) {
        vector<string> res;
        if (digits.empty()) {
            return res;
        }
        std::string combination;
        backtrack(digits, 0, combination, res);
        return res;
    }
};
```

````python
# 使用 Pyhton 求解
from typing import List

class Solution:
    def letterCombinations(self, digits: str) -> List[str]:
        res: List[str] = []
        if digits is None:
            return res

        phone_map = {
            '2': "abc",
            '3': "def",
            '4': "ghi",
            '5': "jkl",
            '6': "mno",
            '7': "pqrs",
            '8': "tuv",
            '9': "wxyz",
        }

        def backtrack(index: int, combination: List[str]):
            if len(combination) == len(digits):
                res.append(''.join(combination))
                return
            
            current_char: str = digits[index]
            letters: str = phone_map[current_char]

            for letter in letters:
                combination.append(letter)
                backtrack(index + 1, combination)
                combination.pop()
            
        backtrack(0, [])
        return res
````
