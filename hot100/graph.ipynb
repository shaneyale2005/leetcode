# [Leetcode200-岛屿数量](https://leetcode.cn/problems/number-of-islands/?envType=study-plan-v2&envId=top-100-liked)

给你一个由'1'（陆地）和'0'（水）组成的的二维网格，请你计算网格中岛屿的数量。

岛屿总是被水包围，并且每座岛屿只能由水平方向和“/”或竖直方向上相邻的陆地连接形成。

此外，你可以假设该网格的四条边均被水包围。

```cpp
// 使用 Cpp 求解
#include <deque>
class Solution {
public:
    int orangesRotting(vector<vector<int>>& grid) {
        // 首先确定参数
        int rows = grid.size();
        int cols = grid[0].size();
        int fresh_count = 0;
        std::deque<vector<int>> dq;
        for (int i = 0; i < rows; i++) {
            for (int j = 0; j < cols; j++) {
                if (grid[i][j] == 1) {
                    fresh_count += 1;
                } else if (grid[i][j] == 2) {
                    dq.push_back({i, j});
                }
            }
        }

        if (fresh_count == 0) {
            return 0;
        }

        // 然后进行 BFS
        int minutes = 0;
        vector<pair<int, int>> directions = {{0, 1}, {0, -1}, {1, 0}, {-1, 0}};
        while (!dq.empty() && fresh_count > 0) {
            minutes += 1;
            int current_size = dq.size();
            for (int i = 0; i < current_size; i++) {
                vector<int> temp = dq.front();
                dq.pop_front();
                for (const auto &dir : directions) {
                    int nr = temp[0] + dir.first;
                    int nc = temp[1] + dir.second;

                    if (nr >= 0 && nr < rows && nc >= 0 && nc < cols && grid[nr][nc] == 1) {
                        grid[nr][nc] = 2;
                        fresh_count -= 1;
                        dq.push_back({nr, nc});
                    }
                }
            }
        }
        if (fresh_count == 0) {
            return minutes;
        } else {
            return -1;
        }
    }
};
```

````python
from typing import List
# 使用 Py 求解
class Solution:
    def nunIslands(self, grid: List[List[str]]) -> int:
        if not grid:
            return 0
        m, n = len(grid), len(grid[0]) # m * n 的矩阵
        count = 0

        def dfs(i: int, j: int) -> None:
            if i < 0 or i >= m or j < 0 or j >= n:
                return
            
            if grid[i][j] == '0':
                return
            
            grid[i][j] = '0'

            dfs(i + 1, j)
            dfs(i, j + 1)
            dfs(i - 1, j)
            dfs(i, j - 1)

        for i in range(m):
            for j in range(n):
                if grid[i][j] == '1':
                    count += 1
                    dfs(i, j)

        return count
````

 # [Leetcode994-腐烂的橘子](https://leetcode.cn/problems/rotting-oranges/description/?envType=study-plan-v2&envId=top-100-liked)

 在给定的mxn网格grid中，每个单元格可以有以下三个值之一。

- 值0代表空单元格；
- 值1代表新鲜橘子；

每分钟，腐烂的橘子周围4个方向上相邻的新鲜橘子都会腐烂。

返回直到单元格中没有新鲜橘子为止所必须经过的最小分钟数。如果不可能，返回-1。

````python
from typing import List
from collections import deque
# 使用 Py 求解
class Solution:
    def orangesRotting(self, grid: List[List[int]]) -> int:
        rows = len(grid)
        cols = len(grid[0])
        queue = deque()
        fresh_count = 0
        # 接下来记录初始的状态
        for i in range(rows):
            for j in range(cols):
                if grid[i][j] == 2:
                    queue.append((i, j))
                elif grid[i][j] == 1:
                    fresh_count += 1
            
        if fresh_count == 0:
            return 0

        minutes = 0
        while queue and fresh_count > 0:
            minutes += 1
            for _ in range(len(queue)):
                i, j = queue.popleft()
                for dr, dc in [(0, 1), [0, -1], [1, 0], [-1, 0]]:
                    nr, nc = i + dr, j + dc

                    # 如果遇到了好橘子，就变坏
                    if 0 <= nr < rows and 0 <= nc < cols and grid[nr][nc] == 1:
                        grid[nr][nc] = 2
                        fresh_count -= 1
                        queue.append((nr, nc))
            
        # 看看最后还有没有好的橘子
        return minutes if fresh_count == 0 else -1
                
                
````

// 使用 Cpp 求解
#include <vector>
#include <queue>

class Solution {
public:
    // 抽象，判断一个图是否是有环的
    bool canFinish (int numCourses, vector<vector<int>> &prerequisities) {
        vector<int> inDegree(numCourses, 0); // 首先统计每门课的入度
        vector<vector<int>> adjacency(numCourses); // 一个邻接表

        for (const auto& pre : prerequisities) {
            adjacency[pre[1]].push_back(pre[0]);
            inDegree[pre[0]]++;
        }

        queue<int> q;
        // 首先把不需要先修课的节点找出来
        for (int i = 0; i < numCourses; i++) {
            if (inDegree[i] == 0) {
                q.push(i);
            }
        }

        int count = 0;
        while (!q.empty()) {
            int curr = q.front();
            q.pop();
            count += 1;
            for (int nextCourse : adjacency[curr]) {
                inDegree[nextCourse]--;
                if (inDegree[nextCourse] == 0) {
                    q.push(nextCourse);
                }
            }
        }
        return count == numCourses;
    }
};
