# [Leetcode 215. 数组中第 K 个最大的元素](https://leetcode.cn/problems/kth-largest-element-in-an-array/description/?envType=study-plan-v2&envId=top-100-liked)

```cpp
class Solution {
public:
    int findKthLargest(vector<int>& nums, int k) {
        std::sort(nums.begin(), nums.end());
        std::reverse(nums.begin(), nums.end());
        return nums[k - 1];
    }
};
```

````python
# Solution in Python
from typing import List

class Solution:
    def findKthLargest(self, nums: List[int], k: int) -> int:
        nums.sort()
        nums = nums[::-1]
        return nums[k - 1]
````

# [Leetcode 347. 前 K 个高频元素](https://leetcode.cn/problems/top-k-frequent-elements/description/?envType=study-plan-v2&envId=top-100-liked)

```cpp
// Solution in C++
#include <vector>
#include <unordered_map>
#include <queue>
#include <utility> // Required for std::pair

class Solution {
public:
    std::vector<int> topKFrequent(std::vector<int>& nums, int k) {
        // Step 1: Count the frequency of each element using a hash map.
        // Key: element from nums, Value: its frequency.
        std::unordered_map<int, int> frequencyMap;
        for (int num : nums) {
            frequencyMap[num]++;
        }

        // Step 2: Use a min-heap to get the top k frequent elements.
        // The C++ Standard Library's priority_queue is a max-heap by default.
        // To use it as a min-heap, we must provide a custom comparator.
        // The lambda 'cmp' defines the min-heap logic:
        // For two pairs (freq1, num1) and (freq2, num2), it returns true if freq1 > freq2.
        // This makes the priority_queue act as a min-heap comparing the frequency (the first element of the pair).
        auto cmp = [](const std::pair<int, int>& lhs, const std::pair<int, int>& rhs) {
            return lhs.first > rhs.first;
        };
        
        // The priority_queue is initialized with:
        // 1. The type of elements being stored (std::pair<int, int>).
        // 2. The underlying container type (std::vector).
        // 3. The comparator type, derived from our lambda using decltype.
        std::priority_queue<std::pair<int, int>, 
                            std::vector<std::pair<int, int>>, 
                            decltype(cmp)> minHeap(cmp);

        // Step 3: Iterate through the frequency map and maintain the heap of size k.
        for (const auto& entry : frequencyMap) {
            // The 'entry' is a pair of (element, frequency).
            // We push {frequency, element} onto the heap. The heap sorts by frequency.
            minHeap.push({entry.second, entry.first});

            // If the heap size exceeds k, we remove the element with the smallest frequency.
            // This is the key to the algorithm's efficiency. We only ever keep k elements.
            if (minHeap.size() > k) {
                minHeap.pop();
            }
        }

        // Step 4: Extract the result from the heap.
        // At this point, the heap contains the top k frequent elements.
        // However, when we pop from a min-heap, we get the smallest element first.
        // To get the result in descending order of frequency, we fill the result vector backwards.
        std::vector<int> result(k);
        for (int i = k - 1; i >= 0; i--) {
            result[i] = minHeap.top().second; // Get the element itself (the 'second' in our pair).
            minHeap.pop();
        }

        return result;
    }
};
```
