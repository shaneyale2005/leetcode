# [Leetcode 215. 数组中第 K 个最大的元素](https://leetcode.cn/problems/kth-largest-element-in-an-array/description/?envType=study-plan-v2&envId=top-100-liked)

```cpp
class Solution {
public:
    int findKthLargest(vector<int>& nums, int k) {
        std::sort(nums.begin(), nums.end());
        std::reverse(nums.begin(), nums.end());
        return nums[k - 1];
    }
};
```

````python
# Solution in Python
from typing import List

class Solution:
    def findKthLargest(self, nums: List[int], k: int) -> int:
        nums.sort()
        nums = nums[::-1]
        return nums[k - 1]
````
