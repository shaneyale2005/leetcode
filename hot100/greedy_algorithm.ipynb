# [Leetcode 121. 买卖股票的最佳时机](https://leetcode.cn/problems/best-time-to-buy-and-sell-stock/description/?envType=study-plan-v2&envId=top-100-liked)

```cpp
// Solution in C++
#include <algorithm>
#include <vector>

class Solution {
public:
    int maxProfit(std::vector<int> &prices) {
        int max_profit = 0;
        int min_price = prices[0];

        for (const int & price : prices) {
            min_price = std::min(price, min_price);
            int current_profit = price - min_price;
            max_profit = std::max(max_profit, current_profit);
        }

        return max_profit;

    }
};
```

````python
# Solution in Python

from typing import List

class Solution:
    def maxProfit(self, prices: List[int]) -> int:
        max_profit: int = 0
        min_price: int = prices[0]

        for price in prices:
            min_price = min(min_price, price)
            current_profit: int = price - min_price
            max_profit = max(max_profit, current_profit)

        return max_profit
````

# [Leetcode 55. 跳跃游戏](https://leetcode.cn/problems/jump-game/?envType=study-plan-v2&envId=top-100-liked)

```cpp
// Solution in C++
#include <vector>
#include <algorithm>

class Solution {
public:
    bool canJump(std::vector<int> &nums) {
        int n = nums.size();
        int max_reach = 0;

        for (int i = 0; i < n; i++) {
            if (max_reach < i) {
                return false;
            }
            max_reach = std::max(max_reach, i + nums[i]);
            if (max_reach >= n - 1) {
                return true;
            }
        }

        return true;
    }
};
```

````python
# Solution in Python
from typing import List

class Solution:
    def canJump(self, nums: List[int]) -> bool:
        n = len(nums)
        max_reach: int = 0 # max_reach 是一种索引
        for i in range(n):
            if i > max_reach:
                return False
            max_reach = max(max_reach, i + nums[i])
            if max_reach >= n - 1:
                return True
        return True
````

# [Leetcode 45. 跳跃游戏 II](https://leetcode.cn/problems/jump-game-ii/?envType=study-plan-v2&envId=top-100-liked)

```cpp
// Solution in C++
class Solution {
public:
    int jump(std::vector<int> &nums) {
        int steps = 0;
        int cur_end = 0;
        int max_far = 0;
        int n = nums.size();

        for (int i = 0; i < n - 1; i++) {
            max_far = std::max(i + nums[i], max_far);
            if (i == cur_end) {
                steps++;
                cur_end = max_far;
                if (cur_end >= n -1) {
                    break;
                }
            }
        }
        return steps;
    }
};

```

````python
# Solution in Python
from typing import List

class Solution:
    def jump(self, nums: List[int]) -> int:
        steps: int = 0
        cur_end: int = 0
        max_far: int = 0
        n: int = len(nums)

        for i in range(0, n - 1):
            max_far = max(i + nums[i], max_far)
            if i == cur_end:
                steps += 1
                cur_end = max_far
                if cur_end >= n - 1:
                    break
        return steps
````

# [Leetcode 763. 划分字母区间](https://leetcode.cn/problems/partition-labels/description/?envType=study-plan-v2&envId=top-100-liked)

```cpp
class Solution {
public:
    vector<int> partitionLabels(string s) {
        int last[26] = {0};
        for (int i = 0; i < s.size(); i++) {
            last[s[i] - 'a'] = i;
        }

        vector<int> ans;
        int start = 0;
        int end = 0;
        for (int i = 0; i < s.size(); i++) {
            end = max(end, last[s[i] - 'a']);
            if (i == end) {
                ans.push_back(end - start + 1);
                start = i + 1;
            }
        }
        return ans;
    }
};
```

````python
# Solution in Python
from typing import List

class Solution:
    def partitionLabels(self, s: str) -> List[int]:

        # 先记录每个字母最后出现的位置
        last = {c : i for i,c in enumerate(s)}
        ans = []
        start = 0
        end = 0
        for i, c in enumerate(s):
            end = max(end, last[c])
            if i == end:
                ans.append(end - start + 1)
                start = i + 1

        return ans
````
