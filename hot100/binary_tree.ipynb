# [Leetcode101-对称二叉树](https://leetcode.cn/problems/symmetric-tree/description/?envType=study-plan-v2&envId=top-100-liked)

给你一个二叉树的根节点root， 检查它是否轴对称。

```cpp
// 使用Cpp进行求解
class Solution {
public:
    bool isMirror(TreeNode *left, TreeNode *right) {
        // 终止条件
        if (left == nullptr && right == nullptr) {
            return true;
        }
        
        // 一个为空，另一个不是空，不对称
        if ((left && right == nullptr) || (right && left == nullptr)) {
            return false;
        }

        // 节点的数值不相等，不对称
        if (left->val != right->val) {
            return false;
        }

        // 递归比较
        return isMirror(left->left, right->right) && isMirror(left->right, right->left);
    }

    bool isSymmetric(TreeNode *root) {
        // 判断一棵树是否是对称的，等价于判断他的左子树和右子树是否是镜像对称的，可以使用递归
        if (!root) { return true; }
        return isMirror(root);
    }
};
```

````python
# 使用Py进行求解
# Definition for a binary tree node.
class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right

class Solution:
    def isSymmetric(self, root: Optional[TreeNode]) -> bool:
        if root is None:
            return True
        def isMirror(left, right):
            # 如果都是空，返回真
            if left is None and right is None:
                return True
            # 如果有一个为空，返回假
            if left and right is None or right and left is None:
                return False
            if left.val != right.val:
                return False
            
            return isMirror(left.left, right.right) and isMirror(left.right, right.left)
        return isMirror(root.left, root.right)
````

# [Leetcode543-二叉树的直径](https://leetcode.cn/problems/diameter-of-binary-tree/?envType=study-plan-v2&envId=top-100-liked)

给你一棵二叉树的根节点，返回该树的直径。

二叉树的直径是指树中任意两个节点之间最长路径的长度。这条路径可能经过也可能不经过根节点root。

两节点之间路径的长度由它们之间边数表示。

```cpp
// 使用Cpp求解
class Solution {
private:
    int ans = 0;
public:
    // 思路：针对任何一个root，经过它的最大直径是左子树的maxDepth加上右子树的maxDepth
    int maxDepth(TreeNode *root) {
        if (root == nullptr) {
            return 0;
        }
        int left = maxDepth(root->left);
        int right = maxDepth(root->right);
        ans = std::max(left + right, ans);
        return std::max(left, right) + 1;
    }
    int diameterOfBinaryTree(TreeNode *root) {
        ans = 0;
        maxDepth(root);
        return ans;
    }
};
```



````python
# 使用Py进行求解
class Solution:
    def diameterOfBinaryTree(self, root):
        self.ans = 0
        def depth(root):
            ans = 0
            if not root:
                return 0
            left = depth(root.left)
            right = depth(root.right)
            self.ans = max(left + right, self.ans)
            return max(left, right) + 1
        ans = depth(root)
        return self.ans
````

# [Leetcode102-二叉树的层序遍历](https://leetcode.cn/problems/binary-tree-level-order-traversal/description/?envType=study-plan-v2&envId=top-100-liked)

给你二叉树的根节点root，返回其节点值的层序遍历。

```cpp
// 使用Cpp进行求解
class Solution {
public:
    vector<vector<int>> levelOrder(TreeNode *root) {
        // 常见的思路就是用队列来做
        vector<vector<int>> res;
        if (!root) {
            return res;
        }
        queue<TreeNode*> q;
        q.push(root);
        while (!q.empty()) {
            int levelSize = q.size();
            vector<int> currentLevel;
            for (int i = 0; i < levelSize; i++) {
                TreeNode *node = q.front();
                q.pop();
                currentLevel.push_back(node->val);
                if (node->left) {
                    q.push(node->left);
                }
                if (node->right) {
                    q.push(node->right);
                }
            }
            res.push_back(currentLevel);
        }
        return res;
    }
};

```

````python
# 使用Py进行求解

from collections import deque

class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = None
        self.right = None
    
class Solution:
    def levelOrder(self, root):
        res = []
        if root is None:
            return res
        
        q = deque()
        q.append(root)
        while q:
            levelSize = len(q)
            currentLevel = []
            for i in range(levelSize):
                node = q.popleft()
                currentLevel.append(node.val)
                if node.left:
                    q.append(node.left)
                if node.right:
                    q.append(node.right)
            res.append(currentLevel)
        return res
````

给你一个整数数组nums，其中元素已经按升序排列，请你将其转换为一棵平衡二叉搜索树。

```cpp

class Solution {
private:
    TreeNode *buildBST(vector<int> &nums, int left, int right) {
        if (left > right) {
            return nullptr;
        }
        int mid = left + (right - left) / 2;
        TreeNode *root = new TreeNode(nums[mid]);
        root->left = buildBST(nums, left, mid - 1);
        root->right = buildBST(nums, mid + 1, right);
        return root;
    }
public:
    TreeNode *sortedArrayToBST(vector<int> &nums) {
        return buildBST(nums, 0, nums.size() - 1);
    }
};
```

````python
# Solution in Python
class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right

class Solution:
    def sortedArrayToBST(self, nums):
        def buildBST(left, right):
            if left > right:
                return None
            mid = left + (right - left) // 2
            root = TreeNode(nums[mid])

            root.left = buildBST(left, mid - 1)
            root.right = buildBST(mid + 1, right)
            return root
        
        return buildBST(0, len(nums) - 1)
````

给定二叉树的根节点root，判断该二叉树是否为有效的二叉搜索树。

有效二叉搜索树的定义如下：
- 节点的左子树中所有节点的值都严格小于当前节点的值；
- 节点的右子树中所有节点的值都严格大于当前节点的值；

 *     int val;
 *     TreeNode *left;
 *     TreeNode *right;
 *     TreeNode() : val(0), left(nullptr), right(nullptr) {}
 *     TreeNode(int x) : val(x), left(nullptr), right(nullptr) {}
 *     TreeNode(int x, TreeNode *left, TreeNode *right) : val(x), left(left), right(right) {}
 * };
 */

// 使用Cpp进行求解
class Solution {
public:
    // 基本思路是做一个中序遍历，如果是递增顺序，正确，不是递增顺序，错误
    bool isValidBST(TreeNode *root) {
        vector<long long> inorder;
        inorderTraversal(root, inorder);

                return false;
            }
        }
        return true;
    }

 *     TreeNode *left;
 *     TreeNode *right;

class Solution {
public:

    // 这里用BFS求解

        vector<int> result;
        if (root == nullptr) return result;
        queue<TreeNode*> que;
        que.push(root);
        // 队列不为空，说明还有要处理

            for (int i = 0; i < levelSize; i++) {
                TreeNode *curNode = que.front();
                que.pop();
                if (i == levelSize - 1) {
                    result.push_back(curNode->val);
                }
                if (curNode->left) {
                    que.push(curNode->left);
                }
                if (curNode->right) {
                    que.push(curNode->right);
                }
            }
        }
        return result;
    }

# [Leetcode105-从前序与中序遍历序列构造二叉树](https://leetcode.cn/problems/construct-binary-tree-from-preorder-and-inorder-traversal/description/?envType=study-plan-v2&envId=top-100-liked)

给定两个整数数组preorder和inorder，其中preorder是二叉树的先序遍历，inorder是同一棵树的中序遍历，请构造二叉树并返回其根节点。

```cpp
// 使用Cpp进行求解
class Solution {
public:
    unordered_map<int, int> index;  // 记录 inorder 中每个值的位置

    /**
    @param: preL和PreR表示当前要处理的子树的先序遍历是什么
    @param: inL和inR表示当前要处理的子树的中序遍历是什么
    */
    TreeNode* build(
        vector<int>& preorder, int preL, int preR,
        vector<int>& inorder,  int inL,  int inR
    ) {
        // 递归的出口
        if (preL > preR) return nullptr;

        // 先序的第一个元素是根
        int rootVal = preorder[preL];
        TreeNode* root = new TreeNode(rootVal);

        // 根在中序中的位置
        int k = index[rootVal];
        int leftSize = k - inL;

        // 构造左右子树
        root->left = build(
            preorder, preL + 1, preL + leftSize,
            inorder,  inL, k - 1
        );

        root->right = build(
            preorder, preL + leftSize + 1, preR,
            inorder,  k + 1, inR
        );

        return root;
    }

    TreeNode* buildTree(vector<int>& preorder, vector<int>& inorder) {
        // 建立中序值到下标的映射
        for (int i = 0; i < inorder.size(); ++i) {
            index[inorder[i]] = i;
        }

        return build(
            preorder, 0, preorder.size() - 1,
            inorder,  0, inorder.size() - 1
        );
    }
};


```

````python
class Solution:
    # 我们将作为哈希表的字典放在 self 上，这样在递归时也能访问到
    def __init__(self):
        # key是中序遍历的节点值, value是该节点在中序数组中的下标
        self.inorder_index_map = {}
    def buildTree(self, preorder: list[int], inorder: list[int]) -> TreeNode | None:
        """
        主函数，用于初始化并启动递归构建过程。
        """
        # 1. 提前构建好中序遍历的值到索引的映射，这是关键优化
        for i, val in enumerate(inorder):
            self.inorder_index_map[val] = i
        # 2. 调用核心递归函数，注意传入整个数组的范围
        #    preL, preR, inL, inR 都是 0 和 len-1
        return self._build_tree_recursive(
            preorder,
            0,
            len(preorder) - 1,
            inorder,
            0,
            len(inorder) - 1
        )
    def _build_tree_recursive(self, preorder, pre_start, pre_end, inorder, in_start, in_end):
        """
        递归构建树的函数。
        pre_start, pre_end: 当前子树在 preorder 数组中的起始和结束索引。
        in_start, in_end: 当前子树在 inorder 数组中的起始和结束索引。
        """
        # 递归的出口：如果子树的范围无效，说明节点为空
        if pre_start > pre_end:
            return None
        # --- 根据分治法思想开始构建 ---
        # 1. 先序遍历的第一个元素必然是当前子树的根
        root_val = preorder[pre_start]
        # print(f"当前处理节点: {root_val}") # <--- 调试语句
        root = TreeNode(root_val)
        # 2. 在中序遍历中找到根节点的位置
        #    复杂度从 O(n) 降为 O(1)
        k = self.inorder_index_map[root_val]
        
        # 3. 计算左子树的大小
        #    k 是根的位置，in_start 是左子树的起始位置，左子树节点数就是 k - in_start
        left_subtree_size = k - in_start
        # print(f"根节点 {root_val} 在中序的位置: {k}, 左子树大小: {left_subtree_size}") # <--- 调试语句
        # 4. 递归地构建左子树和右子树
        # 构建左子树
        # preorder: 根节点是 pre_start，左子树从 pre_start + 1 开始，大小是 left_subtree_size，所以结束于 pre_start + left_subtree_size
        # inorder:  左子树的范围依然是 [in_start, k - 1]
        root.left = self._build_tree_recursive(
            preorder,
            pre_start + 1,
            pre_start + left_subtree_size,
            inorder,
            in_start,
            k - 1
        )
        # 构建右子树
        # preorder: 右子树从左子树后面开始，即 pre_start + left_subtree_size + 1，结束于 pre_end
        # inorder:  右子树的范围是 [k + 1, in_end]
        root.right = self._build_tree_recursive(
            preorder,
            pre_start + left_subtree_size + 1,
            pre_end,
            inorder,
            k + 1,
            in_end
        )
        # 5. 返回当前构建好的树的根节点
        return root
````

# [Leetcode437-路径总和III](https://leetcode.cn/problems/path-sum-iii/description/?envType=study-plan-v2&envId=top-100-liked)


给定一个二叉树的根节点root，和一个整数targetSum，求该二叉树里节点值之和等于targetSum的路径的数目。

```cpp

 * Definition for a binary tree node.
 * struct TreeNode {
 *     int val;
 *     TreeNode *left;
 *     TreeNode *right;
 *     TreeNode() : val(0), left(nullptr), right(nullptr) {}
 *     TreeNode(int x) : val(x), left(nullptr), right(nullptr) {}
 *     TreeNode(int x, TreeNode *left, TreeNode *right) : val(x), left(left), right(right) {}
 * };
 */

// 前缀和
class Solution {
public:
    std::unordered_map<long long, int> prefix_counts;

    int pathSum (TreeNode *root, int targetSum) {
        prefix_counts.clear();
        prefix_counts[0] = 1;
        return dfs(root, 0, targetSum);
    }


private:
    int dfs (TreeNode *node, long long current_sum, int target) {
        if (!node) {
            return 0;
        }

        int count = 0;
        current_sum += node->val;

        // 如果当前前缀和减去目标值的结果在哈希表中，说明找到了符合条件的区间
        if (prefix_counts.count(current_sum - target)) {
            count = prefix_counts[current_sum - target];
        }

        // 更新当前前缀和的频率，递归进入子节点
        prefix_counts[current_sum]++;
        count += dfs(node->left, current_sum, target);
        count += dfs(node->right, current_sum, target);

        prefix_counts[current-sum]--;
        return count;
    }
};
```

````python
# 使用Py求解

# Definition for a binary tree node.
from typing import Optional
from collections import defaultdict

class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right

class Solution:
    def pathSum(self, root: Optional[TreeNode], targetSum: int) -> int:
        prefix_counts = defaultdict(int)

        prefix_counts[0] = 1

        def dfs(node: Optional[TreeNode], current_sum: int) -> int:
            if not node:
                return 0
            
            current_sum += node.val
            required_prefix = current_sum - targetSum
            count = prefix_counts[required_prefix]

            prefix_counts[current_sum] +=1

            count += dfs(node.left, current_sum)
            count += dfs(node.right, current_sum)

            prefix_counts[current_sum] -= 1

            if prefix_counts[current_sum] == 0:
                del prefix_counts[current_sum]

            return count
        return dfs(root, 0)
````

# [Leetcode236-二叉树最近的公共祖先](https://leetcode.cn/problems/lowest-common-ancestor-of-a-binary-tree/?envType=study-plan-v2&envId=top-100-liked)

给定一个二叉树, 找到该树中两个指定节点的最近公共祖先。

```cpp

struct TreeNode {
    int val
    TreeNode *left,
    TreeNode *right,
    TreeNode (int x) : val(x), left(nullpte), right(nullptr) {}
};

// 使用Cpp求解
class Solution {
public:
    TreeNode *lowestCommonAncestor (TreeNode *root, TreeNode *p, TreeNode *q) {
        if (root == nullptr) {
            return nullptr;
        }

        if (root == p || root == q) {
            return root;
        }

        TreeNode *left = lowestCommonAncestor(root->left, p, q);
        TreeNode *right = lowestCommonAncestor(root->right, p, q);

        if (left && right) {
            return root;
        }
        return left ? left : right;
    }
};

```




````python
# 使用Py求解
class TreeNode:
    def __init__(self, x, left, right):
        self.val = x
        self.left = None
        self.right = None
        
class Solution:
    def lowestCommonAncestor(self, root, p, q):
        if root is None:
            return None
        if root is p or root is q:
            return root
        
        left = self.lowestCommonAncestor(root.left, p, q)
        right = self.lowestCommonAncestor(root.right, p, q)

        if left and right:
            return root
        if left:
            return left
        else:
            return right
````

# [Leetcode124-二叉树中的最大路径和](https://leetcode.cn/problems/binary-tree-maximum-path-sum/description/?envType=study-plan-v2&envId=top-100-liked)

二叉树中的路径被定义为一条节点序列，序列中每对相邻节点之间都存在一条边。

同一个节点在一条路径序列中至多出现一次。该路径至少包含一个节点，且不一定经过根节点。

路径和是路径中各节点值的总和。

给你一个二叉树的根节点root，返回其最大路径和。

```cpp
// 使用Cpp求解

#include <limits>
class Solution {
private:
    int ans = INT_MIN;
    int dfs (TreeNode *node) {
        if (node == nullptr) {
            return 0;
        }

        int left_gain = std::max(dfs(node->left), 0);
        int right_gain = std::max(dfs(node->right), 0);

        int current_path = node->val + left_gain + right_gain;
        ans = std::max(current_path, ans);
        return node->val + std::max(left_gain, right_gain);
    }


public:
    int maxPathSum (TreeNode *root) {
        dfs(root);
        return ans;
    }
};
```


````python
# 使用Py求解
from typing import Optional
class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right

class Solution:
    def maxPathSum(self, root: Optional[TreeNode]) -> int:
        self.ans = float('-inf')  # 一个全局
        def dfs(node: Optional[TreeNode]):
            if not node:
                return 0
            
            left_gain = max(dfs(node.left), 0)
            right_gain = max(dfs(node.right), 0)

            current_path = node.val + left_gain + right_gain
            self.ans = max(self.ans, current_path)

            return node.val + max(left_gain, right_gain)
        
        dfs(root)
        return self.ans
    
````
