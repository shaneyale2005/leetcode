# [Leetcode 70. 爬楼梯](https://leetcode.cn/problems/climbing-stairs/description/?envType=study-plan-v2&envId=top-100-liked)

```cpp
// 本质上是斐波那契数列的应用
class Solution {
public:
    int climbStairs(int n) {
        // dp[i] 表示爬到第 i 阶楼梯的方法数，dp[i] = dp[i - 1] + dp[i - 2]
        if (n <= 2) {
            return n;
        }
        vector<int> dp(n + 1);
        dp[0] = 1; // 0 是起点
        dp[1] = 1; // 爬到第 1 阶只有 1 种方法
        for (int i = 2; i <=n; i++) {
            dp[i] = dp[i - 1] + dp[i - 2];
        }
        return dp[n];
    }
};
```

````python
# Solution in Python
from typing import List

class Solution:
    def climbStairs(self, n: int) -> int:
        """
        斐波那契数列的应用
        用 dp[i] 代表爬到第 i 阶的方法数，有递推公式 dp[i] = dp[i - 1] + dp[i - 2]
        """
        
        # 处理边界情况
        if n == 1:
            return 1
        if n == 2:
            return 2
        
        dp: List[int] = [0] * (n + 1)
        dp[0] = 1
        dp[1] = 1
        for i in range(2, n + 1):
            dp[i] = dp[i - 1] + dp[i - 2]
        return dp[n]
````

# [Leetcode 118. 杨辉三角](https://leetcode.cn/problems/pascals-triangle/description/?envType=study-plan-v2&envId=top-100-liked)

```cpp
// Solution in C++
class Solution {
public:
    vector<vector<int>> generate(int numRows) {
        vector<vector<int>> res(numRows);
        for (int i = 0; i < numRows; i++) {
            res[i].resize(i + 1);
            res[i][0] = 1;
            res[i][i] = 1;
            for (int j = 1; j < i; j++) {
                res[i][j] = res[i - 1][j - 1] + res[i - 1][j];
            }
        }
        return res;
    }
};
```

````python
# Solution in Python
from typing import List

class Solution:
    def generate(self, numRows: int) -> List[List[int]]:
        res: List[List[int]] = [[1]]
        for i in range(1, numRows):
            prev_row: List[int] = res[-1]
            curr_row: List[int] = [1]
            for j in range(1, i):
                curr_row.append(prev_row[j - 1] + prev_row[j])
            curr_row.append(1)
            res.append(curr_row)
        return res
````

# [Leetcode 198. 打家劫舍](https://leetcode.cn/problems/house-robber/?envType=study-plan-v2&envId=top-100-liked)

```cpp
// Solution in C++
class Solution {
public:
    /**
     * dp[i] 表示前 i 间房屋能抢到的最大的金额
     * 状态转移公式是 dp[i] = max(dp[i - 1], dp[i - 2] + nums[i])
     */
    int rob(vector<int> &nums) {
        if (nums.empty()) {
            return 0;
        }
        if (nums.size() == 1) {
            return nums[0];
        }

        int prev_prev = nums[0];
        int prev = max(nums[0], nums[1]);

        for (int i = 2; i < nums.size(); i++) {
            int current = max(prev, prev_prev + nums[i]);
            prev_prev = prev;
            prev = current;
        }
        return prev;
    }
};
```

````python
# Solution in Python
from typing import List

class Solution:
    def rob(self, nums: List[int]) -> int:
        if len(nums) == 0:
            return 0
        if len(nums) == 1:
            return nums[0]
        prev_prev: int = nums[0]
        prev: int = max(nums[1], nums[0])
        for i in range(2, len(nums)):
            current: int = max(prev, prev_prev + nums[i])
            prev_prev = prev
            prev = current
        return prev
````

# [Leetcode 279. 完全平方数](https://leetcode.cn/problems/perfect-squares/description/?envType=study-plan-v2&envId=top-100-liked)

```cpp
// Solution in C++
class Solution {
public:
    /**
     * 使用动态规划的方式进行求解
     * dp[i] 表示的是和为 i 的完全平方数的最小的数量
     * 状态转移公式是 dp[i] = min(dp[i], dp[i - j * j] + 1)
     */
    int numSquares(int n) {
        vector<int> dp(n + 1, n);
        dp[0] = 0; // 0 是 0 个完全平方数的和
        for (int i = 1; i <= n; i++) {
            for (int j = 1; j * j <= i; j++) {
                dp[i] = min(dp[i], dp[i - j * j] + 1);
            }
        }
        return dp[n];
    }
};
```

````python
# Solution in Python
from typing import List

class Solution:
    def numSquares(self, n: int) -> int:
        """
        动态规划
        dp[i] 表示的是和为 i 的完全平方数的最小个数
        状态转移公式为 dp[i] = min(dp[i], dp[i - j * j] + 1)
        """
        dp: List[int] = [n] * (n + 1)
        dp[0] = 0
        for i in range(1, n + 1):
            j: int = 1
            while j * j <= i:
                dp[i] = min(dp[i], dp[i - j * j] + 1)
                j += 1
        return dp[n]
````

# [Leetcode 322. 零钱兑换](https://leetcode.cn/problems/pascals-triangle/?envType=study-plan-v2&envId=top-100-liked)

```cpp
// Solution in C++
class Solution {
public:
    /**
     * 动态规划
     * dp[i] 表示凑成金额为 i 需要的最少的硬币的个数
     * 状态转移公式 dp[i] = min(dp[i], dp[i - coin] + 1)
     */
    int coinChange(vector<int> &coins, int amount) {
        vector<int> dp(amount + 1, amount + 1);
        dp[0] = 0; // 凑成金额为 0，需要 0 个硬币
        for (int i = 1; i <= amount; i++) {

                    dp[i] = min(dp[i], dp[i - coin] + 1);
                }
            }
        }

            return -1;
        }
        return dp[amount];
    }

```

````python
# Solution in Python
from typing import List
import math

class Solution:
    def coinChange(self, coins: List[int], amount: int):
        """
        动态规划
        这里的 dp[i] 表示的是凑成金额 i 需要的最少的硬币数量
        状态转移公式 dp[i] = min(dp[i], dp[i - coin] + 1)
        """
        dp = [math.inf] * (amount + 1) # 这里初始化的值都是 amount + 1，因为
        dp[0] = 0 # 凑成金额 0，则需要 0 个硬币，符合直觉
        for i in range(1, amount + 1):
            for coin in coins:
                if i >= coin:
                    dp[i] = min(dp[i], dp[i - coin] + 1)
        if dp[amount] == math.inf:
            return -1
        return dp[amount]
````

# [Leetcode 139. 单词拆分](https://leetcode.cn/problems/word-break/?envType=study-plan-v2&envId=top-100-liked)

```cpp
// Solution in C++
class Solution {
public:
    /**
     * 动态规划
     * 本题中的 dp[i] 表示的是字符串 s 中的前 i 个字符是否能够被成功的拆分
     * 状态转移公式 dp[i] = dp[j] && words.count(s,substr(j, i - j))
     */
    bool wordBreak(string s, vector<string> &wordDict) {
        unordered_set<string> words(wordDict.begin(), wordDict.end());
        int n = s.size();

        vector<bool> dp(n + 1, false);
        dp[0] = true; // 这里是为了让结果能够合法，默认让空串能够拆分
        for (int i = 1; i <= n; i++) {
            for (int j = 0; j < i; j++) {
                if (dp[j] && words.count(s.substr(j, i - j))) {
                    dp[i] = true;
                    break;
                }
            }
        }
        return dp[n];
    }
};
```

````python
# Solution in Python
from typing import List
import math

class Solution:
    def wordBreak(self, s: str, wordDict: List[str]) -> bool:
        """
        动态规划
        这里的 dp[i] 表示的是字符串 s 中的前 i 个字符是否可以被拆分
        状态转移方程是 dp[i] = dp[j] and s[j:i] in words
        """
        n = len(s)
        dp = [False] * (n + 1)
        dp[0] = True
        for i in range(1, n + 1): # 用来遍历 dp 数组
            for j in range(i):
                if dp[j] and s[j:i] in wordDict:
                    dp[i] = True
                    break

        return dp[n]
````

# [Leetcode 300. 最长递增子序列](https://leetcode.cn/problems/longest-increasing-subsequence/?envType=study-plan-v2&envId=top-100-liked)

```cpp
// Solution in C++
class Solution {
public:
    int lengthOfLIS(vector<int> &nums) {
        int n = nums.size();
        if (n == 1) {
            return 1;
        }
        vector<int> dp(n, 1);
        for (int i = 1; i < n; i++) {
            for (int j = 0; j < i; j++) {
                if (nums[j] < nums[i]) {
                    dp[i] = max(dp[i], dp[j] + 1);
                }
            }
        }
        return *max_element(dp.begin(), dp.end());
    }
};
```

````python
# Solution in Python
class Solution:
    def lengthOfLIS(self, nums):
        n = len(nums)
        if n == 1:
            return 1
        dp = [1] * n
        for i in range(1, n): # 遍历 dp
            for j in range(i): # 遍历 i 之前的
                if nums[j] < nums[i]:
                    dp[i] = max(dp[i], dp[j] + 1)
        return max(dp)
````

# [Leetcode 152. 乘积最大子数组](https://leetcode.cn/problems/maximum-product-subarray/description/?envType=study-plan-v2&envId=top-100-liked)

```cpp
// Solution in C++
class Solution {
public:
    /**
     * 动态规划
     *
     * maxDp[i] 表示以 nums[i] 结尾的连续子数组的最大乘积
     * minDp[i] 表示以 nums[i] 结尾的连续子数组的最小乘积
     *
     * 之所以同时维护最大值和最小值，是因为当前元素可能为负数
     * 此时前面的最小乘积乘上当前元素后，可能变成最大的乘积
     *
     * maxDp[i] = max(nums[i], max(maxDp[i - 1] * nums[i], minDp[i - 1] * nums[i]))
     * minDp[i] = min(nums[i], min(maxDp[i - 1] * nums[i], minDp[i - 1] * nums[i]))
     *
     * maxDp[0] = nums[0]
     * minDp[0] = nums[0]
     *
     * 答案为所有 maxDp[i] 中的最大值
     */
    int maxProduct(vector<int> &nums) {
        int n = nums.size();
        vector<int> maxDp(n);
        vector<int> minDp(n);
        maxDp[0] = nums[0];
        minDp[0] = nums[0];
        int ans = nums[0];
        for (int i = 1; i < n; i++) {
            maxDp[i] = max(nums[i], max(maxDp[i - 1] * nums[i], minDp[i - 1] * nums[i]));
            minDp[i] = min(nums[i], min(maxDp[i - 1] * nums[i], minDp[i - 1] * nums[i]));
            ans = max(ans, maxDp[i]);
        }

        return ans;
    }
};
```

````python

class Solution:
    def maxProduct(self, nums):
        """
        动态规划

        maxDp[i] 表示以 nums[i] 结尾的连续子数组的最大乘积
        minDp[i] 表示以 nums[i] 结尾的连续子数组的最小乘积

        之所以同时维护最大值和最小值，
        是因为当前元素可能为负数，
        前面的最小乘积乘上当前元素后，可能变成最大的乘积

        maxDp[i] = max(nums[i], maxDp[i - 1] * nums[i], minDp[i - 1] * nums[i])
        minDp[i] = min(nums[i], maxDp[i - 1] * nums[i], minDp[i - 1] * nums[i])
        maxDp[0] = nums[0]
        minDp[0] = nums[0]

        答案为所有 maxDp[i] 中的最大值
        """
        n = len(nums)
        maxDp = [float("-inf")] * n
        minDp = [float("-inf")] * n
        maxDp[0] = nums[0]
        minDp[0] = nums[0]
        ans = nums[0]
        for i in range(1, n):
            maxDp[i] = max(nums[i], maxDp[i - 1] * nums[i], minDp[i - 1] * nums[i])
            minDp[i] = min(nums[i], maxDp[i - 1] * nums[i], minDp[i - 1] * nums[i])
            ans = max(ans, maxDp[i])
        return ans
````

# [Leetcode 416. 分割等和子集](https://leetcode.cn/problems/partition-equal-subset-sum/description/?envType=study-plan-v2&envId=top-100-liked)

```cpp
// Solution in C++
class Solution {
public:
    bool canPartition(vector<int> &nums) {
        int sum = 0;
        for (int elem : nums) {
            sum += elem;
        }
        if (sum % 2 != 0) {
            return false;
        }
        int target = sum / 2;
        vector<bool> dp(target + 1, false);
        dp[0] = true;
        for (int num : nums) {
            for (int j = target; j >= num; j--) {
                dp[j] = dp[j] || dp[j - num];
            }
        }
        return dp[target];
    }
};
```

````python
# Solution in Python
class Solution:
    def canPartition(self, nums):
        total = sum(nums)
        if total % 2 != 0:
            return False
        target = total // 2
        dp = [False] * (target + 1)
        dp[0] = True
        for num in nums:
            for j in range(target, num - 1, -1):
                dp[j] = dp[j] or dp[j - num]
        return dp[target]
````

# [Leetcode 32. 最长有效括号](https://leetcode.cn/problems/longest-valid-parentheses/description/?envType=study-plan-v2&envId=top-100-liked)

```cpp
// Solution in C++
class Solution {
public:
    int longestValidParentheses(string s) {
        stack<int> st;
        st.push(-1); // 哨兵
        int ans = 0;
        for (int i = 0; i < s.size(); i++) {
            if (s[i] == '(') {
                st.push(i);
            } else {
                st.pop(); // 匹配一个左括号
                if (st.empty()) {
                    st.push(i); // 当前的右括号无效，作为新断点
                } else {
                    ans = max(ans, i - st.top()); // 计算有效长度
                }
            }
        }
        return ans;
    }
};

```

````python
# Solution in Python
class Solution:
    def longestValidParentheses(self, s: str) -> int:
        stack = [-1]
        ans = 0
        for i in range(len(s)):
            if s[i] == '(':
                stack.append(i)
            else:
                stack.pop()
                if not stack:
                    stack.append(i)
                else:
                    ans = max(ans, i - stack[-1])
        return ans
````
