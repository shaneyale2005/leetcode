# [Leetcode 70. 爬楼梯](https://leetcode.cn/problems/climbing-stairs/description/?envType=study-plan-v2&envId=top-100-liked)

```cpp
// 本质上是斐波那契数列的应用
class Solution {
public:
    int climbStairs(int n) {
        // dp[i] 表示爬到第 i 阶楼梯的方法数，dp[i] = dp[i - 1] + dp[i - 2]
        if (n <= 2) {
            return n;
        }
        vector<int> dp(n + 1);
        dp[0] = 1; // 0 是起点
        dp[1] = 1; // 爬到第 1 阶只有 1 种方法
        for (int i = 2; i <=n; i++) {
            dp[i] = dp[i - 1] + dp[i - 2];
        }
        return dp[n];
    }
};
```

````python
# Solution in Python
from typing import List

class Solution:
    def climbStairs(self, n: int) -> int:
        """
        斐波那契数列的应用
        用 dp[i] 代表爬到第 i 阶的方法数，有递推公式 dp[i] = dp[i - 1] + dp[i - 2]
        """
        
        # 处理边界情况
        if n == 1:
            return 1
        if n == 2:
            return 2
        
        dp: List[int] = [0] * (n + 1)
        dp[0] = 1
        dp[1] = 1
        for i in range(2, n + 1):
            dp[i] = dp[i - 1] + dp[i - 2]
        return dp[n]
````
