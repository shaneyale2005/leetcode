# [Leetcode 70. 爬楼梯](https://leetcode.cn/problems/climbing-stairs/description/?envType=study-plan-v2&envId=top-100-liked)

```cpp
// 本质上是斐波那契数列的应用
class Solution {
public:
    int climbStairs(int n) {
        // dp[i] 表示爬到第 i 阶楼梯的方法数，dp[i] = dp[i - 1] + dp[i - 2]
        if (n <= 2) {
            return n;
        }
        vector<int> dp(n + 1);
        dp[0] = 1; // 0 是起点
        dp[1] = 1; // 爬到第 1 阶只有 1 种方法
        for (int i = 2; i <=n; i++) {
            dp[i] = dp[i - 1] + dp[i - 2];
        }
        return dp[n];
    }
};
```

````python
# Solution in Python
from typing import List

class Solution:
    def climbStairs(self, n: int) -> int:
        """
        斐波那契数列的应用
        用 dp[i] 代表爬到第 i 阶的方法数，有递推公式 dp[i] = dp[i - 1] + dp[i - 2]
        """
        
        # 处理边界情况
        if n == 1:
            return 1
        if n == 2:
            return 2
        
        dp: List[int] = [0] * (n + 1)
        dp[0] = 1
        dp[1] = 1
        for i in range(2, n + 1):
            dp[i] = dp[i - 1] + dp[i - 2]
        return dp[n]
````

# [Leetcode 118. 杨辉三角](https://leetcode.cn/problems/pascals-triangle/description/?envType=study-plan-v2&envId=top-100-liked)

```cpp
// Solution in C++
class Solution {
public:
    vector<vector<int>> generate(int numRows) {
        vector<vector<int>> res(numRows);
        for (int i = 0; i < numRows; i++) {
            res[i].resize(i + 1);
            res[i][0] = 1;
            res[i][i] = 1;
            for (int j = 1; j < i; j++) {
                res[i][j] = res[i - 1][j - 1] + res[i - 1][j];
            }
        }
        return res;
    }
};
```

````python
# Solution in Python
from typing import List

class Solution:
    def generate(self, numRows: int) -> List[List[int]]:
        res: List[List[int]] = [[1]]
        for i in range(1, numRows):
            prev_row: List[int] = res[-1]
            curr_row: List[int] = [1]
            for j in range(1, i):
                curr_row.append(prev_row[j - 1] + prev_row[j])
            curr_row.append(1)
            res.append(curr_row)
        return res
````

# [Leetcode 198. 打家劫舍](https://leetcode.cn/problems/house-robber/?envType=study-plan-v2&envId=top-100-liked)

```cpp
// Solution in C++
class Solution {
public:
    /**
     * dp[i] 表示前 i 间房屋能抢到的最大的金额
     * 状态转移公式是 dp[i] = max(dp[i - 1], dp[i - 2] + nums[i])
     */
    int rob(vector<int> &nums) {
        if (nums.empty()) {
            return 0;
        }
        if (nums.size() == 1) {
            return nums[0];
        }

        int prev_prev = nums[0];
        int prev = max(nums[0], nums[1]);

        for (int i = 2; i < nums.size(); i++) {
            int current = max(prev, prev_prev + nums[i]);
            prev_prev = prev;
            prev = current;
        }
        return prev;
    }
};
```

````python
# Solution in Python
from typing import List

class Solution:
    def rob(self, nums: List[int]) -> int:
        if len(nums) == 0:
            return 0
        if len(nums) == 1:
            return nums[0]
        prev_prev: int = nums[0]
        prev: int = max(nums[1], nums[0])
        for i in range(2, len(nums)):
            current: int = max(prev, prev_prev + nums[i])
            prev_prev = prev
            prev = current
        return prev
````

# [Leetcode 279. 完全平方数](https://leetcode.cn/problems/perfect-squares/description/?envType=study-plan-v2&envId=top-100-liked)

```cpp
// Solution in C++
class Solution {
public:
    /**
     * 使用动态规划的方式进行求解
     * dp[i] 表示的是和为 i 的完全平方数的最小的数量
     * 状态转移公式是 dp[i] = min(dp[i], dp[i - j * j] + 1)
     */
    int numSquares(int n) {
        vector<int> dp(n + 1, n);
        dp[0] = 0; // 0 是 0 个完全平方数的和
        for (int i = 1; i <= n; i++) {
            for (int j = 1; j * j <= i; j++) {
                dp[i] = min(dp[i], dp[i - j * j] + 1);
            }
        }
        return dp[n];
    }
};
```

````python
# Solution in Python
from typing import List

class Solution:
    def numSquares(self, n: int) -> int:
        """
        动态规划
        dp[i] 表示的是和为 i 的完全平方数的最小个数
        状态转移公式为 dp[i] = min(dp[i], dp[i - j * j] + 1)
        """
        dp: List[int] = [n] * (n + 1)
        dp[0] = 0
        for i in range(1, n + 1):
            j: int = 1
            while j * j <= i:
                dp[i] = min(dp[i], dp[i - j * j] + 1)
                j += 1
        return dp[n]
````
