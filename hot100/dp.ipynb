# [Leetcode 70. 爬楼梯](https://leetcode.cn/problems/climbing-stairs/description/?envType=study-plan-v2&envId=top-100-liked)

```cpp
// 本质上是斐波那契数列的应用
class Solution {
public:
    int climbStairs(int n) {
        // dp[i] 表示爬到第 i 阶楼梯的方法数，dp[i] = dp[i - 1] + dp[i - 2]
        if (n <= 2) {
            return n;
        }
        vector<int> dp(n + 1);
        dp[0] = 1; // 0 是起点
        dp[1] = 1; // 爬到第 1 阶只有 1 种方法
        for (int i = 2; i <=n; i++) {
            dp[i] = dp[i - 1] + dp[i - 2];
        }
        return dp[n];
    }
};
```

````python
# Solution in Python
from typing import List

class Solution:
    def climbStairs(self, n: int) -> int:
        """
        斐波那契数列的应用
        用 dp[i] 代表爬到第 i 阶的方法数，有递推公式 dp[i] = dp[i - 1] + dp[i - 2]
        """
        
        # 处理边界情况
        if n == 1:
            return 1
        if n == 2:
            return 2
        
        dp: List[int] = [0] * (n + 1)
        dp[0] = 1
        dp[1] = 1
        for i in range(2, n + 1):
            dp[i] = dp[i - 1] + dp[i - 2]
        return dp[n]
````

# [Leetcode 118. 杨辉三角](https://leetcode.cn/problems/pascals-triangle/description/?envType=study-plan-v2&envId=top-100-liked)

```cpp
// Solution in C++
class Solution {
public:
    vector<vector<int>> generate(int numRows) {
        vector<vector<int>> res(numRows);
        for (int i = 0; i < numRows; i++) {
            res[i].resize(i + 1);
            res[i][0] = 1;
            res[i][i] = 1;
            for (int j = 1; j < i; j++) {
                res[i][j] = res[i - 1][j - 1] + res[i - 1][j];
            }
        }
        return res;
    }
};
```

````python
# Solution in Python
from typing import List

class Solution:
    def generate(self, numRows: int) -> List[List[int]]:
        res: List[List[int]] = [[1]]
        for i in range(1, numRows):
            prev_row: List[int] = res[-1]
            curr_row: List[int] = [1]
            for j in range(1, i):
                curr_row.append(prev_row[j - 1] + prev_row[j])
            curr_row.append(1)
            res.append(curr_row)
        return res
````
