# [Leetcode74-搜索二维矩阵](https://leetcode.cn/problems/search-a-2d-matrix/?envType=study-plan-v2&envId=top-100-liked)

```cpp
// Solution in C++
class Solution {
private:
    bool binarySearch (vector<int> &nums, int target) {
        int left = 0;
        int right = nums.size() - 1;

        while (left <= right) {
            int mid = (left + right) / 2;
            if (nums[mid] == target) {
                return true;
            } else if (nums[mid] > target) {
                right = mid - 1; 
            } else {
                left = mid + 1;
            }
        }
        return false;
    }

public:
    bool searchMatrix (vector<vector<int>> &matrix, int target) {
        int m = matrix.size();
        int n = matrix[0].size();
        vector<int> one_dimension;
        for (int i = 0; i < m; i++) {
            for (int j = 0; j < n; j++) {
                one_dimension.push_back(matrix[i][j]);
            }
        }
        bool res = binarySearch(one_dimension, target);
        return res;
    }
};
```

````python
# Solution in Python

from typing import List

class Solution:
    def searchMatrix(self, matrix: List[List[int]], target: int) -> bool:
        """
        主要是降维处理
        """
        m: int = len(matrix)
        n: int = len(matrix[0])
        one_dimension: List[int] = []
        for row in matrix:
            for elem in row:
                one_dimension.append(elem)

        def binary_search(nums: List[int]) -> bool:
            left: int = 0
            right: int = len(nums) - 1

            while left <= right:
                mid: int = (left + right) // 2

                if nums[mid] == target:
                    return True
                elif nums[mid] > target:
                    right = mid - 1
                else:
                    left = mid + 1
            return False

        res: bool = binary_search(one_dimension)
        return res
````

# [Leetcode34-在排序数组中查找元素的第一个和最后一个位置](https://leetcode.cn/problems/find-first-and-last-position-of-element-in-sorted-array/?envType=study-plan-v2&envId=top-100-liked)

```cpp
// Solution in C++

class Solution {
private:
    /**
     * 用来寻找左边界，第一个 >= target 的位置
     */
    int findLeft (vector<int> &nums, int target) {
        int left = 0;
        int right = nums.size() - 1;

        while (left <= right) {
            int mid = (left + right) / 2;
            if (nums[mid] >= target) {
                right = mid - 1; // 向左收缩
            } else {
                left = mid + 1; // 向右收缩
            }
        }
        return left;
    }

    /**
     * 用来寻找右边界，第一个 <= target 的位置
     */
    int findRight (vector<int> &nums, int target) {
        int left = 0;
        int right = nums.size() - 1;

        while (left <= right) {
            int mid = (left + right) / 2;

            if (nums[mid] <= target) {
                left = mid + 1;
            } else {
                right = mid - 1;
            }
        }

        return right;
    }
public:
    vector<int> searchRange (vector<int> &nums, int target) {
        if (nums.empty()) {
            return {-1, -1};
        }

        int left = findLeft(nums, target);
        int right = findRight(nums, target);
        if (left <= right && right < nums.size() && nums[left] == target && nums[right] == target) {
            return {left, right};
        }

        return {-1, -1};
    }

};
```

````python
# Solution in Python

from typing import List

class Solution:
    def searchRange(self, nums: List[int], target: int) -> List[int]:
        if nums is None:
            return [-1, -1]
        
        def find_left(nums: List[int], target: int) -> int:
            """
            用来寻找左边界，第一个大于等于 target 的
            """
            left: int = 0
            right: int = len(nums) - 1
            res: int = -1

            while left <= right:
                mid: int = (left + right) // 2

                if nums[mid] == target:
                    res = mid
                    right = mid - 1
                elif nums[mid] < target:
                    left = mid + 1
                else:
                    right = mid - 1

            return res
        
        def find_right(nums: List[int], target: int) -> int:
            """
            用来寻找右边界，第一个小于等于 target 的
            """
            left: int = 0
            right: int = len(nums) - 1
            res: int = -1

            while left <= right:
                mid: int = (left + right) // 2

                if nums[mid] == target:
                    res = mid
                    left = mid + 1
                elif nums[mid] > target:
                    right = mid - 1
                else:
                    left = mid + 1
            return res
        
        left = find_left(nums=nums, target=target)
        right = find_right(nums=nums, target=target)

        return [left, right]
````

# [Leetcode33-搜索旋转排序数组](https://leetcode.cn/problems/search-in-rotated-sorted-array/description/?envType=study-plan-v2&envId=top-100-liked)

```cpp
// Solution in C++

class Solution {
public:
    int search(vector<int> &nums, int target) {
        int left = 0;
        int right = nums.size() - 1;

        while (left <= right) {
            int mid = (left + right) / 2;

            if (nums[mid] == target) {
                return mid;
            }

            // Check if the left half [left ... mid] is sorted
            if (nums[left] <= nums[mid]) {
                // Target is in the left sorted half
                if (nums[left] <= target && target <= nums[mid]) {
                    right = mid - 1;
                } else {
                    left = mid + 1
                }
            } else {
                // Right half is sorted
                if (nums[mid] <= target && target <= nums[right]) {
                    left = mid + 1;
                } else {
                    right = mid - 1;
                }
            }
        }
        return -1;
    }
};

```

````python
# Solution in Python

from typing import List

class Solution:
    def search(self, nums: List[int], target: int) -> int:
        left: int = 0
        right: int = len(nums) - 1

        while left <= right:
            mid: int = (left + right) // 2

            # In this case, the index mid is exactly what we are looking for
            if nums[mid] == target:
                return mid
            
            # If the left half is sorted
            if nums[mid] >= nums[left]:
                if nums[left] <= target and target <= nums[mid]:
                    right = mid - 1 # Search in the left half
                else:
                    left = mid + 1 # Search in the right half
            else:
                if nums[mid] <= target and target <= nums[right]:
                    left = mid + 1 # Search in the right half
                else:
                    right = mid - 1 # Search in the left half
        return -1
````

# [Leetcode153-寻找旋转排序数组中的最小值](https://leetcode.cn/problems/find-minimum-in-rotated-sorted-array/?envType=study-plan-v2&envId=top-100-liked)



```cpp
class Solution {
public:
    int findMin(vector<int>& nums) {
        int left = 0;
        int right = nums.size() - 1;
        
        // Binary search for the minimum element
        while (left < right) {
            // Avoid overflow: (left + right) might overflow for large values
            int mid = left + (right - left) / 2;
            
            // Compare mid element with rightmost element
            if (nums[mid] > nums[right]) {
                // Left part is sorted, minimum is in the right part
                // Since nums[mid] > nums[right], mid cannot be the minimum
                left = mid + 1;
            } else {
                // Right part is sorted (including case where mid == right)
                // Minimum could be at mid or in the left part
                // We can't exclude mid because it might be the minimum
                right = mid;
            }
        }
        
        // When left == right, we've found the minimum element
        return nums[left];
    }
};
```

````python
# Solution in Python
from typing import List
class Solution:
    def findMin(self, nums: List[int]) -> int:
        left: int = 0
        right: int = len(nums) - 1

        while left < right:
            mid: int = (left + right) // 2

            if nums[mid] > nums[right]:
                left = mid + 1
            else:
                right = mid

        return nums[left]
````

# [Leetcode4-寻找两个正序数组的中位数](https://leetcode.cn/problems/median-of-two-sorted-arrays/description/?envType=study-plan-v2&envId=top-100-liked)

```cpp
class Solution {
public:
    double findMedianSortedArrays(vector<int> &nums1, vector<int> &nums2) {
        int total = nums1.size() + nums2.size();
        nums1.insert(nums1.end(), nums2.begin(), nums2.end());
        std::sort(nums1.begin(), nums1.end());
        if (total % 2 == 0) {
            // total is an even number
            return (nums1[total / 2] * 1.0 + nums1[total / 2 - 1] * 1.0) / 2.0;
        } else {
            return nums1[total / 2];
        }
    }
};
```

```rust
// Solution in Rust
impl Solution {
    pub fn find_median_sorted_arrays(nums1: Vec<i32>, nums2: Vec<i32>) -> f64 {
        let mut merged = nums1;
        merged.extend(nums2);
        merged.sort_unstable();
        let len = merged.len();
        if len == 0 {
            return 0.0;
        }
        if len % 2 == 0 {
            let mid = len / 2;
            (merged[mid - 1] as f64 + merged[mid] as f64) / 2.0
        } else {
            merged[len / 2] as f64
        }
    }
}
```

````python
# Solution in Python
from typing import List

class Solution:
    def findMedianSortedArrays(self, nums1: List[int], nums2: List[int]) -> float:
        total: int = len(nums1) + len(nums2)
        nums1.extend(nums2)
        nums1 = sorted(nums1)

        if total % 2 == 0:
            return (nums1[total // 2] + nums1[total // 2 - 1]) / 2
        else:
            return nums1[total // 2]
````
