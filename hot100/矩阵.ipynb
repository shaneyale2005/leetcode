# [Leetcode73-矩阵置零](https://leetcode.cn/problems/set-matrix-zeroes/description/?envType=study-plan-v2&envId=top-100-liked)

给定一个mxn的矩阵，如果一个元素为0，则将其所在行和列的所有元素都设为0。

请使用原地算法。

```cpp
// 使用Cpp进行求解
class Solution {
public:
    // 注意是用额外的空间来记录需要清零的行和列
    // 注意是使用原地的算法
    void setZeroes (vector<vector<int>> &matrix) {
        int m = matrix.size();
        int n = matrix[0].size();
        vector<bool> rowZero(m, false);
        vector<bool> colZero(n, false);
        for (int i = 0; i < m; i++) {
            for (int j = 0; j < n; j++) {
                if (matrix[i][j] == 0) {
                    rowZero[i] = true;
                    colZero[j] = true;
                }
            }
        }

        for (int i = 0; i < m; i++) {
            for (int j = 0; j < n; j++) {
                if (rowZero[i] || colZero[j]) {
                    matrix[i][j] = 0;
                }
            }
        }
    }
};
```

````python
# 使用Py进行求解
class Solution:
    def setZeroes(self, matrix):
        m = len(matrix)
        n = len(matrix[0])
        rowZero = [0] * m
        colZero = [0] * n
        for i in range(m):
            for j in range(n):
                if matrix[i][j] == 0:
                    rowZero[i] = 1
                    colZero[j] = 1
        for i in range(m):
            for j in range(n):
                if rowZero[i] == 1 or colZero[j] == 1:
                    matrix[i][j] = 0
````

# [Leetcode54-螺旋矩阵](https://leetcode.cn/problems/spiral-matrix/description/?envType=study-plan-v2&envId=top-100-liked)

给你一个`m`行`n`列的矩阵`matrix`，请按照顺时针螺旋顺序，返回矩阵中的所有元素。

```cpp
// 使用Cpp的解法
class Solution {
public:
    vector<int> spiralOrder (vector<vector<int>> & matrix) {
        // 模拟求解
        if (matrix.empty() || matrix[0].empty()) return {};
        // 这里定义需要
        int m = matrix.size();
        int n = matrix[0].size();
        vector<int> res;

        int left = 0;
        int right = n - 1;
        int top = 0;
        int bottom = m - 1;

        while (left <= right && top <= bottom) {
            // 从左到右
            for (int i = left; i <= right; i++) {
                res.push_back(matrix[top][i]);
            }
            top += 1;
            // 从上到下
            for (int j = top; j <= bottom; j++) {
                res.push_back(matrix[j][right]);
            }
            right -= 1;
            // 从右到左
            if (top <= bottom) {
                for (int i = right; i >= left; i--) {
                    res.push_back(matrix[bottom][i]);
                }
                bottom -= 1;
            }
            // 从下到上
            if (left <= right) {
                for (int j = bottom; j >= top; j--) {
                    res.push_back(matrix[j][left]);
                }
                left += 1;
            }
        }
        return res;
    } 
};
```

````python
# 使用Py进行求解
class Solution:
    def spiralOrder(self, matrix):
        if len(matrix) == 0 or len(matrix[0]) == 0:
            return []
        
        m = len(matrix)
        n = len(matrix[0])
        res = []

        left = 0
        right = n - 1
        top = 0
        bottom = m - 1

        while left <= right and top <= bottom:
            # 从左到右遍历
            for i in range(left, right + 1):
                res.append(matrix[top][i])
            top += 1
            # 从上到下遍历
            for j in range(top, bottom + 1):
                res.append(matrix[j][right])
            right -= 1
            # 从右到左遍历
            if top <= bottom:
                for i in range(right, left - 1, -1):
                    res.append(matrix[bottom][i])
                bottom -= 1
            # 从下到上遍历
            if left <= right:
                for j in range(bottom, top - 1, -1):
                    res.append(matrix[j][left])
                left += 1
        return res
````

# [Leetcode48-旋转图像](https://leetcode.cn/problems/rotate-image/description/?envType=study-plan-v2&envId=top-100-liked)

给定一个n×n的二维矩阵matrix表示一个图像。请你将图像顺时针旋转90度。

你必须在原地旋转图像，这意味着你需要直接修改输入的二维矩阵。请不要使用另一个矩阵来旋转图像。

```cpp
// 使用Cpp的解法
class Solution {
public:
    void rotate(vector<vector<int>> &matrix){
        int m = matrix.size();
        int n = matrix[0].size();
        for (int i = 0; i < m; i++){
            for (int j = i + 1; j < n; j++){
                std::swap(matrix[i][j], matrix[j][i]);
            }
        }
        for(int i = 0; i < m; i++) {
            std::reverse(matrix[i].begin(), matrix[i].end());
        }
    }
};
```

````python
# 使用Py进行求解
class Solution:
    def rotate(self, matrix):
        m = len(matrix)
        n = len(matrix[0])
        for i in range(m):
            for j in range(i+1, n):
                matrix[i][j], matrix[j][i] = matrix[j][i], matrix[i][j]

        for i in range(m):
            matrix[i].reverse()
````
