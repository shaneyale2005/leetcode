# [Leetcode73-矩阵置零](https://leetcode.cn/problems/set-matrix-zeroes/description/?envType=study-plan-v2&envId=top-100-liked)

给定一个mxn的矩阵，如果一个元素为0，则将其所在行和列的所有元素都设为0。

请使用原地算法。

```cpp
// 使用Cpp进行求解
class Solution {
public:
    // 注意是用额外的空间来记录需要清零的行和列
    // 注意是使用原地的算法
    void setZeroes (vector<vector<int>> &matrix) {
        int m = matrix.size();
        int n = matrix[0].size();
        vector<bool> rowZero(m, false);
        vector<bool> colZero(n, false);
        for (int i = 0; i < m; i++) {
            for (int j = 0; j < n; j++) {
                if (matrix[i][j] == 0) {
                    rowZero[i] = true;
                    colZero[j] = true;
                }
            }
        }

        for (int i = 0; i < m; i++) {
            for (int j = 0; j < n; j++) {
                if (rowZero[i] || colZero[j]) {
                    matrix[i][j] = 0;
                }
            }
        }
    }
};
```

````python
# 使用Py进行求解
class Solution:
    def setZeroes(self, matrix):
        m = len(matrix)
        n = len(matrix[0])
        rowZero = [0] * m
        colZero = [0] * n
        for i in range(m):
            for j in range(n):
                if matrix[i][j] == 0:
                    rowZero[i] = 1
                    colZero[j] = 1
        for i in range(m):
            for j in range(n):
                if rowZero[i] == 1 or colZero[j] == 1:
                    matrix[i][j] = 0
````
