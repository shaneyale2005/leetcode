# [Leetcode3-无重复字符的最长子串](https://leetcode.cn/problems/longest-substring-without-repeating-characters/description/?envType=study-plan-v2&envId=top-100-liked)

给定一个字符串 s ，请你找出其中不含有重复字符的最长子串的长度。 

示例 1:
- 输入: s = "abcabcbb"
- 输出: 3 
- 解释: 因为无重复字符的最长子串是 "abc"，所以其长度为 3。

示例 2:
- 输入: s = "bbbbb"
- 输出: 1
- 解释: 因为无重复字符的最长子串是 "b"，所以其长度为 1。

示例 3:
- 输入: s = "pwwkew"
- 输出: 3
- 解释: 因为无重复字符的最长子串是 "wke"，所以其长度为 3。请注意，你的答案必须是子串的长度，"pwke" 是一个子序列，不是子串。
 

提示：
- 0 <= s.length <= $5 * 10^4$
- s 由英文字母、数字、符号和空格组成


```cpp
// 使用Cpp的解法
class Solution{
public:
    int lengthOfLongestSubstring(string s){
        // 注意这里的滑动窗口是左闭右开的写法
        int left = 0;
        int right = 0;
        int res = 0;
        int n = s.size();
        unordered_map<char, int> window;
        while (right < n) {
            char c = s[right];
            if (window.find(c) != window.end() && window[c] >= left) {
                // 注意这里是收缩到重复出现的下一个字符
                left = window[c] + 1;
            }
            window[c] = right;
            ++right;
            res = std::max(res, right - left);
        }
        return res;
    }
};
```

````python
# 使用Py进行求解
class Solution:
    def lengthOfLongestSubstring(self, s):
        left, right, res = 0, 0, 0
        n = len(s)
        window = {}
        while right < n:
            c = s[right]
            if c in window and window[c] >= left:
                # 还是注意这里的更新条件
                left = window[c] + 1
            window[c] = right
            right += 1
            res = max(res, right - left)
        return res
````

# [Leetcode438-找到字符串中所有字母异位词](https://leetcode.cn/problems/find-all-anagrams-in-a-string/description/?envType=study-plan-v2&envId=top-100-liked)

给定两个字符串 s 和 p，找到 s 中所有 p 的 异位词 的子串，返回这些子串的起始索引。不考虑答案输出的顺序。

示例 1:
- 输入: s = "cbaebabacd", p = "abc"
- 输出: [0,6]
- 解释:起始索引等于 0 的子串是 "cba", 它是 "abc" 的异位词。起始索引等于 6 的子串是 "bac", 它是 "abc" 的异位词。

示例 2:
- 输入: s = "abab", p = "ab"
- 输出: [0,1,2]
- 解释:起始索引等于 0 的子串是 "ab", 它是 "ab" 的异位词。起始索引等于 1 的子串是 "ba", 它是 "ab" 的异位词。起始索引等于 2 的子串是 "ab", 它是 "ab" 的异位词。

提示:
- 1 <= s.length, p.length <= $3 * 10^4$
- s 和 p 仅包含小写字母

```cpp
// 使用Cpp的解法
class Solution{
public:
    vector<int> findAnagrams (string s, string p) {
        vector<int> res;
        int n1 = s.size();
        int n2 = p.size();
        if (n1 < n2) {
            return res;
        }
        // 这里用了数组，没用unordered_map
        vector<int> count_p(26, 0);
        vector<int> count_window(26, 0);
        // 初始化，从0开始索引
        for (const auto & elem : p) {
            ++count_p[elem - 'a'];
        }
        for (int i = 0; i < len_p; i++) {
            ++count_window[s[i] - 'a'];
        }
        if (count_p == count_window) {
            res.push_back(0);
        }
        // 这里才是滑动窗口算法的主体部分
        for (int i = 1; i < n1 - n2 + 1; i++) {
            // 把左边的滑出去
            --count_window[s[i - 1] - 'a'];
            ++count_window[s[i + n2 - 1] - 'a'];
            if (count_window == count_p) {
                res.push_back(i);
            }
        }
        return res;
    }   
};
```

````python
class Solution:
    def findAnagrams(self, s, p):
        res = []
        n1 = len(s)
        n2 = len(p)
        if n1 < n2:
            return res
        count_p = [0] * 26
        count_window = [0] * 26
        # 注意Py中的ord()函数，把单个字符转成它的数字
        for c in p:
            count_p[ord(c) - ord('a')] += 1
        for i in range(n2):
            count_window[ord(s[i]) - ord('a')] += 1
        if count_window == count_p:
            res.append(0)
        for i in range(1, n1 - n2 + 1):
            count_window[ord(s[i - 1]) - ord('a')] -= 1
            count_window[ord(s[i + n2 - 1]) - ord('a')] += 1
            if count_p == count_window:
                res.append(i)
        return res
````
