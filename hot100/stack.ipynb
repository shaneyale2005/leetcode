```cpp
// Solution in C++
#include <stack>
class Solution {
public:
    bool isValid(string s) {
        std::stack<char> my_stack;
        for (char ch : s) {
            if (ch == '(' || ch == '[' || ch == '{') {
                my_stack.push(ch);
            } else {
                if (my_stack.empty()) {
                    return false;
                }
                if(ch == ')') {
                    if (my_stack.top() == '(') {
                        my_stack.pop();
                        continue;
                    } else {
                        return false;
                    }
                }

                if (ch == ']') {
                    if (my_stack.top() == '[') {
                        my_stack.pop();
                        continue;
                    } else {
                        return false;
                    }
                }

                if (ch == '}') {
                    if (my_stack.top() == '{') {
                        my_stack.pop();
                        continue;
                    } else {
                        return false;
                    }
                }
            }
        }
        if (my_stack.empty()) {
            return true;
        }
        return false;
    }
};
```

```rust
// Solution in Rust
impl Solution {
    pub fn is_valid(s: String) -> bool {
        let mut stack: Vec<char> = Vec::new();

        for ch in s.chars() {
            match ch {
                '(' | '[' | '{' => stack.push(ch),
                ')' => {
                    if stack.pop() != Some('(') {
                        return false;
                    }
                }
                ']' => {
                    if stack.pop() != Some('[') {
                        return false;
                    }
                }
                '}' => {
                    if stack.pop() != Some('{') {
                        return false;
                    }
                }
                _ => return false,
            }
        }
        stack.is_empty()
    }
}
```

````python
# Solution in Python
from typing import List

class Solution:
    def isValid(self, s: str) -> bool:
        stack: List[str] = []

        for ch in s:
            if ch == '(' or ch == '[' or ch == '{':
                stack.append(ch)
            else:
                if len(stack) == 0:
                    return False
                
                if ch == ')':
                    if stack[-1] == '(':
                        stack.pop()
                        continue
                    else:
                        return False

                if ch == ']':
                    if stack[-1] == '[':
                        stack.pop()
                        continue
                    else:
                        return False
                    
                if ch == '}':
                    if stack[-1] == '{':
                        stack.pop()
                        continue
                    else:
                        return False
        if len(stack) == 0:
            return True
        return False
````

# [Leetcode155-最小栈](https://leetcode.cn/problems/min-stack/?envType=study-plan-v2&envId=top-100-liked)

```cpp
#include <stack>
#include <algorithm>

class Solution {
private:
    struct Node {
        int val;
        int minVal;
        Node(int v, int m) : val(v), minVal(m) {}
    };

    std::stack<Node> s;

public:
    MinStack() {

    }

    void push(int val) {
        if (s.empty()) {
            s.push(Node(val, val));
        } else {
            s.push(Node(val, std::min(val, s.top().minVal)));
        }

    }

    void pop() {
        s.pop();
    }

    int top() {
        return s.top().val;
    }

    int getMin() {
        return s.top().minVal;
    }
};

```

````python
class MinStack:
    def __init__(self):
        self.stack = []
        
    def push(self, val: int) -> None:
        if not self.stack:
            self.stack.append((val, val))
        else:
            current_min = self.stack[-1][1]
            self.stack.append((val, min(val, current_min)))

    def pop(self) -> None:
        self.stack.pop()

    def top(self) -> int:
        return self.stack[-1][0]

    def getMin(self) -> int:
        return self.stack[-1][1]
````

# [Leetcode 394. 字符串解码](https://leetcode.cn/problems/decode-string/description/?envType=study-plan-v2&envId=top-100-liked)

```cpp
// Solution in C++
#include <stack>
#include <utility>
#include <string>
#include <cctype>

using namespace std;

class Solution {
public:
    string decodeString(
        string s
    ) {
        stack<pair<string, int>> stk;
        string res;
        int k = 0;
        for (char c : s) {
            if (isdigit(c)) {
                k = k * 10 + (c - '0');
            } else if (c == '[') {
                stk.emplace(res, k);
                res.clear();
                k = 0;
            } else if (c == ']') {
                auto [last_str, last_k] = stk.top();
                stk.pop();

                string temp;
                temp.reserve(last_k * res.size());
                for (size_t i = 0; i < last_k; i++) {
                    temp += res;
                }
                res = last_str + temp;
            } else {
                res += c;
            }
        }
        return res;
    }
};
```
