# [Leetcode20-有效的括号](https://leetcode.cn/problems/valid-parentheses/?envType=study-plan-v2&envId=top-100-liked)

```cpp
// Solution in C++
#include <stack>
class Solution {
public:
    bool isValid(string s) {
        std::stack<char> my_stack;
        for (char ch : s) {
            if (ch == '(' || ch == '[' || ch == '{') {
                my_stack.push(ch);
            } else {
                if (my_stack.empty()) {
                    return false;
                }
                if(ch == ')') {
                    if (my_stack.top() == '(') {
                        my_stack.pop();
                        continue;
                    } else {
                        return false;
                    }
                }

                if (ch == ']') {
                    if (my_stack.top() == '[') {
                        my_stack.pop();
                        continue;
                    } else {
                        return false;
                    }
                }

                if (ch == '}') {
                    if (my_stack.top() == '{') {
                        my_stack.pop();
                        continue;
                    } else {
                        return false;
                    }
                }
            }
        }
        if (my_stack.empty()) {
            return true;
        }
        return false;
    }
};
```

```rust
// Solution in Rust
impl Solution {
    pub fn is_valid(s: String) -> bool {
        let mut stack: Vec<char> = Vec::new();

        for ch in s.chars() {
            match ch {
                '(' | '[' | '{' => stack.push(ch),
                ')' => {
                    if stack.pop() != Some('(') {
                        return false;
                    }
                }
                ']' => {
                    if stack.pop() != Some('[') {
                        return false;
                    }
                }
                '}' => {
                    if stack.pop() != Some('{') {
                        return false;
                    }
                }
                _ => return false,
            }
        }
        stack.is_empty()
    }
}
```

````python
# Solution in Python
from typing import List

class Solution:
    def isValid(self, s: str) -> bool:
        stack: List[str] = []

        for ch in s:
            if ch == '(' or ch == '[' or ch == '{':
                stack.append(ch)
            else:
                if len(stack) == 0:
                    return False
                
                if ch == ')':
                    if stack[-1] == '(':
                        stack.pop()
                        continue
                    else:
                        return False

                if ch == ']':
                    if stack[-1] == '[':
                        stack.pop()
                        continue
                    else:
                        return False
                    
                if ch == '}':
                    if stack[-1] == '{':
                        stack.pop()
                        continue
                    else:
                        return False
        if len(stack) == 0:
            return True
        return False
````
